```python
import numpy as np 
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
x = [0,1,2]
y = [100, 200, 300]
```

```python
# Plot simple graph
plt.plot(x,y)
plt.show() # Add plt.show() for py script
```

```python
housing = pd.DataFrame({
    'rooms': [1, 1, 2, 2, 2, 3, 3, 3],
    'price': [100, 120, 190, 200, 230, 310, 330, 305]
})
```

```python
housing
```

```python
# Draw scatter plot
plt.scatter(housing['rooms'], housing['price'])
plt.show()
```

```python
plt.plot(x,y, color='red', marker='x', markersize=5, linestyle='--')

# Setup limit axis. Cool stuff
plt.xlim(0,2)
plt.ylim(0,300)

# Style show
plt.title('Simple Graph', color='blue')
plt.xlabel('X Label')
plt.ylabel('Y Label')
```

```python
plt.scatter(housing['rooms'], housing['price'], color='red')

plt.xlim(0,4)
plt.ylim(0, 350)

plt.title('Scatter graph of housing in Malaysia', color='blue')
plt.xlabel('Number of rooms')
plt.ylabel('Price of houses (RM)')

plt.show()
```
