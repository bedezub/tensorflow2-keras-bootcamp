```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart.csv')
```

```python
import seaborn as sns
sns.__version__
```

```python
plt.figure(figsize=(15,10))
sns.distplot(df['age'], kde=False, bins=10, color='red')
```

```python
df.head()
```

```python

sns.countplot(x='cp', data=df, hue='sex')
```

```python
# Box Plot - Middle 50 points. Box aka Whisker Plot.
sns.boxplot(x='target', y='thalach', data=df, hue='sex')
```

```python
sns.scatterplot(x='chol', y='trestbps', data=df, hue='sex', palette='Dark2', size='age')
```
